```python
# outstate and room.board show a strong correlation
# there is also a positive correlation between outstate and grad rate
fig, ax = plt.subplots(ncols=2, figsize=(16, 8))
ax[0].set_title("Outstate vs Graduation Rate")
ax[1].set_title("Room.Board vs Graduation Rate")
college.plot.scatter("Outstate", "Grad.Rate",ax=ax[0]);
college.plot.scatter("Room.Board", "Grad.Rate",ax=ax[1]);
```

```python
# there are very few samples with top10perc > 60
# colleges with top10perc in the 60-100 range have a low acceptance ratio
# as the top10perc increases, there is a decrease in S.F.Ratio
# there is a much clearer decrease in S.F.Ratio as Top25perc increases
college_temp = college.copy()
college_temp["Acceptance_Ratio"] = college["Accept"] / college["Enroll"]
fig, ax = plt.subplots(2, 2, figsize=(16, 16))
ax[0][0].set_title("Top10perc vs Acceptance_Ratio")
ax[0][1].set_title("Top10perc vs S.F.Ratio")
ax[1][0].set_title("Top25perc vs Acceptance_Ratio")
ax[1][1].set_title("Top25perc vs S.F.Ratio")
college_temp.plot.scatter("Top10perc", "Acceptance_Ratio",ax=ax[0][0]);
college_temp.plot.scatter("Top10perc", "S.F.Ratio",ax=ax[0][1]);
college_temp.plot.scatter("Top25perc", "Acceptance_Ratio",ax=ax[1][0]);
college_temp.plot.scatter("Top25perc", "S.F.Ratio",ax=ax[1][1]);
```

```python
# ratio of enrolled students to accepted students
college_temp["Enrollment_ratio"] = college_temp["Enroll"] / college_temp["Accept"]
college_temp.hist("Enrollment_ratio")
```

```python
# as outstate and Room.Board increases, enrollment ratio decreases
fig, ax = plt.subplots(ncols=2, figsize=(16, 8))
ax[0].set_title("Outstate vs Enrollment Ratio")
ax[1].set_title("Room.Board vs Enrollment Ratio")
college_temp.plot.scatter("Outstate", "Enrollment_ratio",ax=ax[0]);
college_temp.plot.scatter("Room.Board", "Enrollment_ratio",ax=ax[1]);
```

```python
# Private colleges tend to have higher boarding and outstate costs
fig, ax = plt.subplots(figsize=(8, 8))
ax.set_title("Costs with Private and Public colleges")
colors = {"Yes":"red", "No":"blue"}
college_temp.plot.scatter("Outstate", "Room.Board",ax=ax, c=college_temp["Private"].map(colors));
```

# 10 (a)

```python
from ISLP import load_data
Boston = load_data('Boston')
Boston.columns
```

# (b)

```python
# 506 rows, 13 columns
# each row represents attributes of a specific house in boston
Boston.shape
```

Column descriptions
- crim: per capita crime rate by town.

- zn: proportion of residential land zoned for lots over 25,000 sq.ft.

- indus: proportion of non-retail business acres per town.

- chas: Charles River dummy variable (= 1 if tract bounds river; 0 otherwise).

- nox: nitrogen oxides concentration (parts per 10 million).

- rm: average number of rooms per dwelling.

- age: proportion of owner-occupied units built prior to 1940.

- dis: weighted mean of distances to five Boston employment centres.

- rad: index of accessibility to radial highways.

- tax: full-value property-tax rate per $10,000.

- ptratio: pupil-teacher ratio by town.

- lstat: lower status of the population (percent).

- medv: median value of owner-occupied homes in $1000s.

# (c)

```python
Boston.hist(figsize=(15, 10));
```

```python
# plotting a subset for better visualization
Boston_subset = Boston[Boston.crim<50]
fig, ax = plt.subplots(figsize=(15, 15))
pd.plotting.scatter_matrix(Boston_subset, ax=ax);
```

```python
# medv vs dis, rad, tax, lstat
fig, ax = plt.subplots(2, 2, figsize=(16, 9))
ax[0][0].set_title("medv vs rm")
ax[0][1].set_title("medv vs lstat")
ax[1][0].set_title("lstat vs rm")
ax[1][1].set_title("dis vs nox")
Boston.plot.scatter("medv", "rm",ax=ax[0][0]);
Boston.plot.scatter("medv", "lstat",ax=ax[0][1]);
Boston.plot.scatter("lstat", "rm",ax=ax[1][0]);
Boston.plot.scatter("dis", "nox",ax=ax[1][1]);
```

Insights:
- As the number of rooms increases, the median value also increases.
- As lstat percent increases, there is a decrease in median value.
- As lstat increases, there is a decrease in the number of rooms.
- As dis increases, nox decreases

# (d)

```python
c = Boston.corr(numeric_only=True)
c = c.sort_values("crim", ascending=False)
c.style.background_gradient(cmap='coolwarm')
```

rad, tax, and lstat are associated with crim.
- rad has a strong positive correlation with crim
- tax has a moderate positive correlation with crim
- lstat has a moderate positive correlation with crim
